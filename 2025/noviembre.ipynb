# 16. Rectangle Count

Given two positive integers representing the width and height of a rectangle, determine how many rectangles can fit in the given one.

Only count rectangles with integer width and height.

For example, given 1 and 3, return 6. Three 1x1 rectangles, two 1x2 rectangles, and one 1x3 rectangle.


```python
def count_rectangles(width, height):
    try:
        w = int(width)
        h = int(height)
    except Exception:
        return 0
    if w <= 0 or h <= 0:
        return 0
    return (w * (w + 1) // 2) * (h * (h + 1) // 2)

count_rectangles(11,19)
```

# 17. Fingerprint Test

Given two strings representing fingerprints, determine if they are a match using the following rules:

- Each fingerprint will consist only of lowercase letters (a-z).
- Two fingerprints are considered a match if:
  - They are the same length.
  - The number of differing characters does not exceed 10% of the fingerprint length.


```python
def is_match(fingerprint_a: str, fingerprint_b: str) -> bool:
    if len(fingerprint_a) != len(fingerprint_b):
        return False

    # Cuenta cuántas posiciones tienen caracteres distintos.
    # zip empareja ambos strings carácter por carácter.
    #   Ejemplo: "hola" y "hila" → zip → (h,h), (o,i), (l,l), (a,a)
    # Luego evalúa a != b en cada par:
    #   (h,h)→False, (o,i)→True, (l,l)→False, (a,a)→False
    # sum(...) convierte True en 1 y False en 0, y suma las diferencias.
    differences = sum(a != b for a, b in zip(fingerprint_a, fingerprint_b))

    return differences <= len(fingerprint_a) / 10

is_match("holamundo1", "holamundo2")
```

# 19. Markdown Heading Converter

**Given a string representing a Markdown heading, return the equivalent HTML heading.**

A valid Markdown heading must:

- Start with zero or more spaces, followed by  
- **1 to 6 hash characters (`#`) in a row**, then  
- **At least one space**, and finally  
- **The heading text**.

The number of hash symbols determines the heading level.  
For example:
- One hash symbol → `<h1>`
- Six hash symbols → `<h6>`

If the given string doesn't have the exact format above, return **"Invalid format"**.

**Example:**  
Given `# My level 1 heading`  
Return:  
`<h1>My level 1 heading</h1>`

*Note: The console may not display HTML tags in strings when logging messages.  
Check the browser console to see logs with tags included.*


```python
def convert(heading: str) -> str:
    heading = heading.lstrip()

    hashes = 0
    for c in heading:
        if c == "#":
            hashes += 1
        else:
            break

    if hashes == 0 or hashes > 6:
        return "Invalid format"

    # Lo que sigue después de los hashes debe comenzar con un espacio
    rest = heading[hashes:]
    if not rest.startswith(" "):
        return "Invalid format"

    # Extraer el texto (sin el espacio inicial)
    text = rest.strip()
    if not text:
        return "Invalid format"

    return f"<h{hashes}>{text}</h{hashes}>"


convert("##### Hola")
```

# 21. LCM
Given two integers, return the least common multiple (LCM) of the two numbers.

The LCM of two numbers is the smallest positive integer that is a multiple of both numbers. For example, given 4 and 6, return 12 because:

- Multiples of 4 are 4, 8, 12 and so on.
- Multplies of 6 are 6, 12, 18 and so on.
- 12 is the smallest number that is a multiple of both.

```python
def gcd(a: int, b: int) -> int:
    a, b = abs(a), abs(b)
    while b:
        a, b = b, a % b
    return a

def lcm(a: int, b: int) -> int:
    if a == 0 and b == 0:
        return 0
    g = gcd(a, b)
    return abs((a // g) * b)

lcm(4,6)
```

# 22. Recipe Scaler


Given an array of recipe ingredients and a number to scale the recipe, return an array with the quantities scaled accordingly.

- Each item in the given array will be a string in the format: "quantity unit ingredient". For example "2 C Flour".
- Scale the quantity by the given number. Do not include any trailing zeros and do not convert any units.
- Return the scaled items in the same order they are given.


```python
def scale_recipe(ingredients, scale):
    def scale_item(item):
        amount, rest = item.split(" ", 1)
        new_amount = float(amount) * scale
        new_amount = int(new_amount) if new_amount.is_integer() else new_amount
        return f"{new_amount} {rest}"
    
    return [scale_item(item) for item in ingredients]

scale_recipe(["4 L agua", "3 C aceite", "6 A harina"], 2.4)
```

# 23. Character Count
Given a sentence string, return an array with a count of each character in alphabetical order.

- Treat upper and lowercase letters as the same letter when counting.
- Ignore numbers, spaces, punctuation, etc.
- Return the count and letter in the format "letter count". For instance, "a 3".
- All returned letters should be lowercase.
- Do not return a count of letters that are not in the given string.


```python
def count_characters(sentence: str) -> list:
    def clean(text: str) -> str:
        return "".join(c for c in text if c.isalpha())

    letters = sorted(clean(sentence.lower()))

    result = []
    if not letters:
        return result

    current = letters[0]
    count = 0

    for c in letters:
        if c == current:
            count += 1
        else:
            result.append(f"{current} {count}")
            current = c
            count = 1

    result.append(f"{current} {count}")

    return result


count_characters("Hoooola!")
```

# 25. FizzBuzz
Given an integer (n), return an array of integers from 1 to n (inclusive), replacing numbers that are multiple of:

- 3 with "Fizz".
- 5 with "Buzz".
- 3 and 5 with "FizzBuzz".

```python
def fizz_buzz(n: int) -> list:
    return [
        ("FizzBuzz" if i % 15 == 0
         else "Fizz" if i % 3 == 0
         else "Buzz" if i % 5 == 0
         else i)
        for i in range(1, n + 1)
    ]


fizz_buzz(10)
```
