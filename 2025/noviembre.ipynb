```python
def generate_signature(name, title, company):
    name_split = name.split()
    first_letter = name_split[0][0].upper()  # primera letra en mayúscula

    if first_letter in 'ABCDEFGHI':
        return f">>{name}, {title} at {company}"
    elif first_letter in 'JKLMNÑOPQR':
        return f"--{name}, {title} at {company}"
    elif first_letter in 'STUVWXYZ':
        return f"::{name}, {title} at {company}"

generate_signature("Gustavo Magallanes", "Profesor de Asignatura", "UNAM")
```

```python
def shift_array(arr, k):
    n = len(arr)
    if n == 0:
        return arr
    k = k % n
    return arr[k:] + arr[:k]
shift_array([0,1,2,3,4,5,6,7], -9)
```

# 16. Rectangle Count

Given two positive integers representing the width and height of a rectangle, determine how many rectangles can fit in the given one.

Only count rectangles with integer width and height.

For example, given 1 and 3, return 6. Three 1x1 rectangles, two 1x2 rectangles, and one 1x3 rectangle.


```python
def count_rectangles(width, height):
    try:
        w = int(width)
        h = int(height)
    except Exception:
        return 0
    if w <= 0 or h <= 0:
        return 0
    return (w * (w + 1) // 2) * (h * (h + 1) // 2)

count_rectangles(11,19)
```
